```python
import numpy as np
import pandas as pd

scores=np.random.randint(40, high=101, size=(50,5))
scores_pd=pd.DataFrame(scores,
    columns=['國','英','數','地','歷'],
    index=range(1,51))
scores_pd.index.name='學號'
scores_pd.columns.name='科目'
sum_series=scores_pd.sum(axis=1)
mean_series=scores_pd.mean(axis=1)
#display(sum_series)
rank_series=sum_series.rank(method='min',ascending=False)
scores_pd['總分']=sum_series
scores_pd['平均']=mean_series
scores_pd['名次']=rank_series
scores_pd.sort_values(by='名次')
```
